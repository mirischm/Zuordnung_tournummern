```python
# Überblick: Wie viele Zeilen bleiben nach dem Merge ohne Tour?
print(f"Gesamt ohne Tournummer: {df_ist_merged['TOURNR'].isna().sum()}")

# Aufschlüsselung: Echte Datenlücken (kein Soll vorhanden) vs. mögliche Zuordnungsfehler
keine_tour = df_ist_merged[df_ist_merged['TOURNR'].isna()]

# Gibt es für den Fahrer+Datum irgendeine Tour in df_soll (unabhängig vom Kennzeichen)?
soll_fd = df_soll.groupby(['FAHRERNAME', 'DATUM'])['TOURNR'].first().reset_index()
soll_fd.columns = ['Fahrername', 'DATUM', 'TOURNR_fallback']

keine_tour_check = keine_tour[['Fahrername', 'DATUM']].drop_duplicates() \
    .merge(soll_fd, on=['Fahrername', 'DATUM'], how='left')

print("\nMit Fallback (nur Fahrer+Datum, ohne Kennzeichen):")
print(keine_tour_check['TOURNR_fallback'].isna().value_counts())
# False = es gibt eine Tour wenn man Kennzeichen ignoriert
# True  = echter Datenfehler, kein Soll vorhanden
```

```python
# Fallback-Mapping nur Fahrer+Datum
soll_fd = df_soll.groupby(['FAHRERNAME', 'DATUM'])['TOURNR'] \
    .first().reset_index()
soll_fd.columns = ['Fahrername', 'DATUM', 'TOURNR_fallback']

df_ist_merged = df_ist_merged.merge(soll_fd, on=['Fahrername', 'DATUM'], how='left')

# Fallback nur dort eintragen wo TOURNR noch NaN ist
df_ist_merged['TOURNR'] = df_ist_merged['TOURNR'].fillna(df_ist_merged['TOURNR_fallback'])
df_ist_merged = df_ist_merged.drop(columns=['TOURNR_fallback'])

# Kontrolle
print(df_ist_merged['TOURNR'].isna().sum(), "Zeilen ohne Tournummer (echte Lücken)")
print(df_ist_merged['TOURNR'].notna().sum(), "Zeilen mit Tournummer")
```

```python
# Welche Fahrer+Datum haben gar keine Soll-Daten?
echte_luecken = df_ist_merged[df_ist_merged['TOURNR'].isna()] \
    .groupby(['Fahrername', 'KENNZ_CLEAN', 'DATUM']).size().reset_index()
echte_luecken.columns = ['Fahrername', 'KENNZ_CLEAN', 'DATUM', 'Anzahl_Meldungen']
print(echte_luecken.to_string(index=False))
```
